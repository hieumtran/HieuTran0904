```python
def get_payoff(alpha, beta, gamma, rho):
    return [[0, alpha, 0], 
            [1 + alpha - beta, 1 - 2 * beta, 1 - beta + rho], 
            [1 - gamma, 1 - gamma, 1 - gamma]]

parameter_values = [[2], [1], [1], [1]]
labels = ['S', 'D', 'I']
simplex = plot_static(parameter_values, custom_func=get_payoff, vert_labels=labels)
```

```python
simplex = plot_static(parameter_values, custom_func=get_payoff, vert_labels=labels,
                        paths=True)
```

```python
simplex = plot_static(parameter_values, custom_func=get_payoff, vert_labels=labels,
                        background=True)
```

```python
simplex = plot_static(parameter_values, custom_func=get_payoff, vert_labels=labels,
                        paths=True, 
                        generations=10,
                        steps=2000,
                        ic_type='random',
                        path_color='viridis',
                        eq=False)
```

```python
def func(A,B,C):
  return A + C - B - (A-2*B)/10

for A in range(1, 10):
  for C in range(1, 10):
    for B in range(0, 10): 
      res = func(A,B,C)
      if res == 0:
        print("A: " + str(A) + " B: " + str(B) + " C: " + str(C))
  
```

```python
i = 0
while func(8,4,i) != 0:
  i+= 1
  print(func(8,4,i))
  if i > 10:
    break
print(i)
```
